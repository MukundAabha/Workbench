## Univariate, Bivariate, Multivariate Data Analysis

When it comes to the level of Data Analysis there are three main different analysis techniques that exist. These are:

1. Univariate Data analysis
2. Bivariate Data analysis
3. Multivariate Data analysis

This is where we analyse ONLY ONE VARIABLE

Univariate analysis is the most basic form of statistical data analysis technique.

When the data contains only one variable and doesnâ€™t deal with causes or effect relationships then a Univariate analysis technique is used. For example we may want to see the distribution of salary in an organisation. In this case we are only interested in the salary not salary and say year of experience or anything else.

The key objective of Univariate analysis is to simply describe the data to find patterns within the data. This is be done by looking into the mean, median, mode, dispersion, variance, range, standard deviation etc.

How Univariate analysis is conducted

Univariate analysis is conducted through several ways which are mostly descriptive in nature

1. Frequency Distribution Tables
2. Histograms
3. Frequency Polygons
4. Pie Charts
5. Bar Charts

### Bivariate Data Analysis

This is where we analyse TWO VARIABLES.

Bivariate analysis is slightly more analytical than Univariate analysis. When the data set contains two variables and researchers aim to undertake comparisons between the two data set then Bivariate analysis is the right type of analysis technique.

For instance, we can analyse the salary employees and their satisfaction level.

How Bivariate analysis is conducted

1. Correlation coefficients: to examine the strength of relationship between two variables
2. Regression analysis: for estimating the relationships between two different variables.

##### Is there any Relationship between employeess Salary and Satisfaction Level?

```python
# In this case we can use the Seaborns's joinplot() function to display the relationship

sns.jointplot(df['average_montly_hours'], df['satisfaction_level'])
```

for this particular data, there is no relationship between employee's satisfaction level and average monthly hours

### Multivariate Data Analysis

This is where we analyse MORE THAN TWO VARIABLES.

Multivariate analysis is a more complex form of statistical analysis technique and used when there are more than two variables in the data set.

For exmple we may want to analyse the satisfaction level, average monthly hours and the salary of the various employess all at a go. Here, we use multivariate data analysis to achieve that.

Let's analyse the Salary and Number of Project and also whether the employee left of not.

In this case we will be dealing with 3 variables: Salary, Number of Project and left

```python
sns.barplot(x="number_project",y="salary",data=df , hue='left')
plt.show()
```

The employees with low and medium salary turn to leave their projects on the way, while employees with high salary have the tendency of not leaving their projects.

1==leave

0==Not leave

```python
sns.pairplot(df,diag_kind='hist',kind='scatter')
```

In the above diagram, we dont have much correlation between the variables.

```python
sns.pairplot(df[['number_project','salary','satisfaction_level']], kind = 'scatter')
plt.show()
```
